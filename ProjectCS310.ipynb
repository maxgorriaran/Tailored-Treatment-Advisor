Diabetes is a disease that occurs when your blood glucose, also called blood sugar, is too high.
Glucose is your body’s main source of energy. Your body can make glucose, but glucose also comes from the food you eat.

Insulin is a hormone made by the pancreas that helps glucose get into your cells to be used for energy. If you have diabetes, your body doesn’t make enough—or any—insulin, or doesn’t use insulin properly. Glucose then stays in your blood and doesn’t reach your cells.

Diabetes raises the risk for damage to the eyes, kidneys, nerves, and heart. Diabetes is also linked to some types of cancer. Taking steps to prevent or manage diabetes may lower your risk of developing diabetes health problems.

# Type 1 diabetes
When your body makes little or no insulin. Your immune system attacks and destroys the cells in your pancreas that make insulin. Type 1 diabetes is usually diagnosed in children and young adults, although it can appear at any age. People with type 1 diabetes need to take insulin every day to stay alive.

# Type 2 diabetes
When the cells in your body don’t use insulin properly. The pancreas may be making insulin but is not making enough insulin to keep your blood glucose level in the normal range. Type 2 diabetes is the most common type of diabetes. You are more likely to develop type 2 diabetes if you have risk factors, such as overweight or obesity, and a family history of the disease. You can develop type 2 diabetes at any age, even during childhood.

You can help delay or prevent type 2 diabetes by knowing the risk factors and taking steps toward a healthier lifestyle, such as losing weight or preventing weight gain

# Gestational diabetes
Gestational diabetes is a type of diabetes that develops during pregnancy. Most of the time, this type of diabetes goes away after the baby is born. However, if you’ve had gestational diabetes, you have a higher chance of developing type 2 diabetes later in life. Sometimes diabetes diagnosed during pregnancy is type 2 diabetes.

# Prediabetes
People with prediabetes have blood glucose levels that are higher than normal but not high enough to be diagnosed with type 2 diabetes. If you have prediabetes, you have a higher risk of developing type 2 diabetes in the future. You also have a higher risk for heart disease than people with normal glucose levels.

this  column  descrie a Plasma glucose concentration 2 hours in an oral glucose tolerance test (OGTT)", the glucose values represent the 2-hour glucose levels from a standard OGTT. This type of test is specifically designed to assess how well the body processes glucose, making it distinct from fasting glucose or random glucose:

Thresholds for the 2-Hour OGTT (Using Plasma Glucose Levels)

Normal: Less than 140 mg/dL

Prediabetes: Between 140–199 mg/dL (Impaired Glucose Tolerance)

Diabetes: Greater than or equal to 200 mg/dL
; 
Note that if in 2-Hour OGTT patient has > 200 mg/dL is not necessary meaning you have diabe, Doctors may confirm diabetes with additional tests, such as: Fasting Plasma Glucose (≥ 126 mg/dL); , Random Plasma Glucose (≥ 200 mg/dL) with symptoms of diabetes (like excessive thirst, urination, or fatigue).Hemoglobin A1c (≥ 6.5%), Random Plasma Glucose (≥ 200 mg/dL) with symptoms of diabetes (like excessive thirst, urination, or fatigue).

# Insulin: 2-hour serum insulin (mu U/ml)

Insulin levels reflect the body's ability to regulate blood glucose. Low insulin production (Type 1 diabetes) or insulin resistance (Type 2 diabetes) are fundamental causes of diabetes. Te insulin can be classified in :

Normal range: 18–174 μU/mL after 2 hours of glucose intake

Low insulin levels (hypoinsulinemia): Below 18 μU/mL: Often associated with Type 1 diabetes or advanced Type 2 diabetes where beta cells are damaged

High insulin levels (hyperinsulinemia): Above 174 μU/mL. Indicates insulin resistance (common in prediabetes and early Type 2 diabetes).

Mean Imputation: Replace missing values with the mean of the column. This method works well when the data is normally distributed.
Median Imputation: Replace missing values with the median of the column. This is more robust to outliers.
Mode Imputation: Used for categorical data, this method replaces missing values with the mode (the most frequent value) of the column.
Interpolation: For numerical or time-series data, this technique estimates missing values based on the surrounding values

# Collaborative Filtering

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

Preprocess Data for Collaborative Filtering

```python
# List of columns to use for collaborative filtering
cf_features = ['Pregnancies', 'Age', 'Glucose_Category', 'BloodPressure_Category', 
               'SkinThickness_Category', 'Insulin_Category', 'BMI_Category', 
               'categorize_dpf', 'DiabetesPedigreeFunction_Category']

# Extract relevant features from the dataset
cf_data = filtered_data[cf_features + ['Outcome']]

# Check for any missing values and handle them (if necessary)
cf_data = cf_data.dropna()  # Drop rows with missing values for simplicity

```

```python
# Compute the pairwise cosine similarity
similarity_matrix = cosine_similarity(cf_data[cf_features])

# Convert similarity matrix into a DataFrame for readability
similarity_df = pd.DataFrame(similarity_matrix, 
                             index=cf_data.index, 
                             columns=cf_data.index)
print("Patient Similarity Matrix:")
print(similarity_df.head())
```

```python
def get_nearest_neighbors(patient_id, k=5):
    # Sort the similarity values for the given patient
    neighbors = similarity_df.loc[patient_id].sort_values(ascending=False)
    
    # Exclude the patient itself (similarity = 1)
    nearest_neighbors = neighbors.iloc[1:k+1]
    return nearest_neighbors

# Example: Find 5 nearest neighbors for patient with index 0
patient_id = 0
k = 5
neighbors = get_nearest_neighbors(patient_id, k)
print(f"Top {k} Nearest Neighbors for Patient {patient_id}:")
print(neighbors)
```

```python
def recommend_outcome(patient_id, k=5):
    # Find nearest neighbors
    neighbors = get_nearest_neighbors(patient_id, k)
    
    # Get their outcomes
    neighbor_outcomes = cf_data.loc[neighbors.index, 'Outcome']
    
    # Predict the outcome based on majority voting
    recommended_outcome = neighbor_outcomes.mode()[0]
    
    return recommended_outcome, neighbor_outcomes

# Example: Recommend for patient 0
recommended_outcome, neighbor_outcomes = recommend_outcome(patient_id, k)
print(f"Recommended Outcome for Patient {patient_id}: {recommended_outcome}")
print(f"Outcomes of Nearest Neighbors: {neighbor_outcomes.values}")
```
